```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Đọc ảnh
image_path = 'C:/bai tap it/hinhanh/hinh1.JPG'
image = Image.open(image_path)
# Ảnh xám
gray_image = image.convert("L")

# Chuyển ảnh xám về mảng
image_a = np.array(gray_image)\

gradient_x = ndimage.sobel(image_a, axis=0, mode="constant")

# Hiển thị ảnh
fig, axis = plt.subplots(2,1, figsize=(10,8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám gốc")

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title("Ảnh gradient_x")

plt.show()
```
